```python
from osgeo import gdal
import matplotlib.pyplot as plt
import numpy as np
  
  
```

```python
dataset = gdal.Open(r'data.tif')
```

```python
print(dataset.RasterCount)
```

```python
#https://www.geeksforgeeks.org/visualizing-tiff-file-using-matplotlib-and-gdal-using-python/
```

```python
band1 = dataset.GetRasterBand(1) 
band2 = dataset.GetRasterBand(2) 
band3 = dataset.GetRasterBand(3)
band4 = dataset.GetRasterBand(4)
```

```python

b1 = band1.ReadAsArray()
b2 = band2.ReadAsArray()
b3 = band3.ReadAsArray()
b4 = band4.ReadAsArray()
```

```python
img = np.dstack((b1, b2, b3))
f = plt.figure()
plt.imshow(img)
plt.savefig('Tiff.png')
plt.show()
```

```python
i = b4
f = plt.figure()
plt.imshow(i)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from skimage.color import rgb2gray
from skimage.filters import sobel
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float


 
segments_slic = slic(img, n_segments=250, compactness=10, sigma=1,
                     start_label=1)
 
print(f'SLIC number of segments: {len(np.unique(segments_slic))}')


fig, ax = plt.subplots(figsize=(20, 20), sharex=True, sharey=True)
 
ax.imshow(mark_boundaries(img, segments_slic))
ax.set_title('SLIC')
 
plt.show()
```

```python

f = plt.figure()
plt.imshow(segments_slic)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=114, random_state=0).fit(segments_slic)
 
```

```python
label = kmeans.labels_
print(label)
```
